```r
contrast_matrix = dba.show(boca2_norm, bContrasts = TRUE)
contrast_matrix
```

```r
granges_to_csv <- function(granges_obj, contrast){
    report <- dba.report(granges_obj, contrast = contrast)
    report_df <- as.data.frame(report)
    region= paste(report_df$seqnames,report_df$start,sep=":")
    region=paste(region,report_df$end,sep="-")
    report_df["region"]=region
    save_name = paste( gsub("-", "_", contrast_matrix[contrast,2], fixed = TRUE) , "_VS_" ,gsub("-", "_", contrast_matrix[contrast,4], fixed = TRUE), ".csv",sep="")
    save_path_full = paste("/project/scATAC_analysis/10X_PBMC_analysis/AnnotationQuestion/overlap_bulk_and_scATAC/BOCA2_differential_cCREs/results/",save_name,sep="")
    write.table( x = report_df, file = save_path_full, sep=",", col.names=TRUE, row.names=FALSE, quote=FALSE )
}
```

```r
for (i in seq(1, NROW(contrast_matrix))){
    granges_to_csv(boca2_norm, i)
    print(i)
}
```

```r
granges_to_bed <- function(granges_obj, contrast){
    report <- dba.report(granges_obj, contrast = contrast)
    bed_df <- data.frame(seqnames=seqnames(report),
    starts=start(report)-1,
    ends=end(report),
    strands=strand(report),
    fold_change=report$Fold,
    p_value=report$`p-value`,
    FDR=report$FDR,
    mean_read_concentration=mcols(report)[1],
    read_concentration_condition1=mcols(report)[2],
    read_concentration_condition2=mcols(report)[3])
    save_name = paste( gsub("-", "_", contrast_matrix[contrast,2], fixed = TRUE) , "_VS_" ,gsub("-", "_", contrast_matrix[contrast,4], fixed = TRUE), ".bed",sep="")
    save_path_full = paste("/project/scATAC_analysis/10X_PBMC_analysis/AnnotationQuestion/overlap_bulk_and_scATAC/BOCA2_differential_cCREs/results/bedFiles/",save_name,sep="")
    write.table(bed_df, file=save_path_full, quote=F, sep="\t", row.names=F, col.names=F)

    
}
```

```r
for (i in seq(1, NROW(contrast_matrix))){
    granges_to_bed(boca2_norm, i)
    print(i)
}
```

```r
granges_to_df <- function(granges_obj, contrast){
    report <- dba.report(granges_obj, contrast = contrast)
    report_df <- as.data.frame(report)
    region= paste(report_df$seqnames,report_df$start,sep=":")
    region=paste(region,report_df$end,sep="-")
    report_df["region"]=region
    return(report_df)
}
```

```r
for (i in seq(1, NROW(contrast_matrix))){
    save_name = paste( gsub("-", "_", contrast_matrix[i,2], fixed = TRUE) , "_VS_" ,gsub("-", "_", contrast_matrix[i,4],  fixed = TRUE), "_df",sep="")
    
    print(save_name)
}
```

```r
glutamatergic_neurons_VS_GABAergic_neurons_df = granges_to_df(boca2_norm, contrast = 1)
glutamatergic_neurons_VS_oligodendrocytes_df = granges_to_df(boca2_norm, contrast = 2)
glutamatergic_neurons_VS_microglia_and_astrocytes_df = granges_to_df(boca2_norm, contrast = 3)
GABAergic_neurons_VS_oligodendrocytes_df = granges_to_df(boca2_norm, contrast = 4)
GABAergic_neurons_VS_microglia_and_astrocytes_df = granges_to_df(boca2_norm, contrast = 5)
microglia_and_astrocytes_VS_oligodendrocytes_df = granges_to_df(boca2_norm, contrast = 6)

```

## FILTER THE DIFFERENTIAL REGIONS:

### GET TOP 2000 REGIONS OF EACH PAIRWISE COMPARISON

##### remove the Y columns!!

```r
combined_regions_top2000 =c()
for (df_ in list(glutamatergic_neurons_VS_GABAergic_neurons_df, glutamatergic_neurons_VS_oligodendrocytes_df,glutamatergic_neurons_VS_microglia_and_astrocytes_df,GABAergic_neurons_VS_oligodendrocytes_df,GABAergic_neurons_VS_microglia_and_astrocytes_df,microglia_and_astrocytes_VS_oligodendrocytes_df)){
    df_= df_[df_[["seqnames"]] !="chrY",]
    top2000_regions = df_[sort(abs(df_[["Fold"]]),decreasing=T,index.return=T)[[2]],][1:2000,][["region"]]
    top2000_merged_name = gsub(":", "_", gsub("-", "_", top2000_regions, fixed = TRUE) , fixed = TRUE)
    combined_regions_top2000 =  c(combined_regions_top2000,top2000_merged_name )
}
```

```r
combined_regions_top2000[1:50]
```

```r
combined_regions_top2000_sorted = sort(combined_regions_top2000)
```

```r
combined_regions_top2000_3col = as.data.frame(combined_regions_top2000_sorted) %>% 
   separate(combined_regions_top2000_sorted, into = c("seq", "start", "end"), sep="_")
```
