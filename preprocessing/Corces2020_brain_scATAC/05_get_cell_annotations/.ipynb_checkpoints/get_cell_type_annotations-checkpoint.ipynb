Supplementary Data Set 2 - Page 1 of 3								
This page of the table shows the quality control information and cluster classifications for each individual single cell.								
								
Row Explanation								
Each row represents a different individual single cell in our scATAC-seq data.								
								
Column Explanations								
Donor_ID - A numeric identifier unique to each individual donor.								
Region - The brain region corresponding to this piece of tissue. SMTG = Supperior and middle temporal gyri; PARL = parietal lobe; MDFG = middle frontal gyrus; CAUD = caudate nucleus; PTMN = putamen; HIPP = hippocampus; SUNI = substantia nigra								
10x_SingleCell_Barcode - the 16-bp unique cellular identification barcode from 10x Genomics for this cell.								
Total_Reads - The total number of reads from the given cell.								
Total_Fragments - The total number of paired-end fragmetns from the given cell.								
Reads_In_Peaks - The total number of reads in peaks from the given cell.								
FRiP - The fraction of reads in peaks from the merged scATAC-seq peak set for the given cell.								
TSS_Enrichment - The TSS enrichment score for the given cell.								
UMAP_dim1 - The coordinate value for the given cell along the UMAP dimension 1.								
UMAP_dim2 - The coordinate value for the given cell along the UMAP dimension 2.								
Cluster - The assigned cluster number for the given cell.								
NeuronCluster - The assigned neuron cluster from neuronal sub-clustering analysis.								
is_Dopaminergic? - A boolean value indicating the cells that were defined as dopaminergic neurons using sub-clustering of substantia nigra samples.								

```python
CAUD_neuronal_cluster_df = dfs['scATAC-seq QC Metadata'].iloc[21:,]
CAUD_neuronal_cluster_df.columns = dfs['scATAC-seq QC Metadata'].iloc[20,]

CAUD_neuronal_cluster_df = CAUD_neuronal_cluster_df[CAUD_neuronal_cluster_df.Region=='CAUD']
CAUD_06_0615_neuronal_cluster_df = CAUD_neuronal_cluster_df[CAUD_neuronal_cluster_df.Donor_ID== '06_0615']
```

```python
CAUD_06_0615_neuronal_cluster_df
```

```python
clust_ID_annotation_dict = dict(zip(cluster_residence_df.Cluster , cluster_residence_df.Cell_Type_Group))

clust_ID_annotation_dict
```

```python
CAUD_06_0615_neuronal_cluster_df_cell_type_annotations = CAUD_06_0615_neuronal_cluster_df[["10x_SingleCell_Barcode", 'Cluster']]
```

```python
CAUD_06_0615_neuronal_cluster_df_cell_type_annotations
```

```python
CAUD_06_0615_neuronal_cluster_df_cell_type_annotations['annotations'] = CAUD_06_0615_neuronal_cluster_df_cell_type_annotations['Cluster'].map(clust_ID_annotation_dict)
```

```python
CAUD_06_0615_neuronal_cluster_df_cell_type_annotations
```

```python
CAUD_06_0615_neuronal_cluster_df_cell_type_annotations["cell_IDs"] = CAUD_06_0615_neuronal_cluster_df_cell_type_annotations["10x_SingleCell_Barcode"] + "-1"
```
